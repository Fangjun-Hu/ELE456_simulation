```python
from qutip import *
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
w_q = 2 * np.pi * 7850.3
chi = 2 * np.pi * 9.4
w_c = 2 * np.pi * 9274.7
K =  2 * np.pi * 0.325
kappa = 2 * np.pi * 10e-3

N = 15    # number of cavity fock states
psi0 = coherent(N, 2)
a = destroy(N)

H = - K/2 * a.dag() * a * a.dag() * a

tlist_1 = np.linspace(0,8,80000)    # time list for simulation
output_1 = mesolve(H, psi0, tlist_1, [np.sqrt(kappa)*a])
```

## no decay

```python
w_q = 2 * np.pi * 7850.3
chi = 2 * np.pi * 9.4
w_c = 2 * np.pi * 9274.7
K   = 2 * np.pi * 0.325
kappa = 2 * np.pi * 10e-3

N = 15    # number of cavity fock states
psi0 = coherent(N, 2)
a = destroy(N)

H = - K/2 * a.dag() * a * a.dag() * a

tlist_2 = np.linspace(0,8,80000)
output_2 = mesolve(H, psi0, tlist_2)
```

# For movie

```python
tot_frames = 50
for index_output in np.arange(0, tot_frames):
    index_output_data = int(index_output * 80000/tot_frames)
    time_point = 8000 * index_output/tot_frames    # ns
    xvec = np.linspace(-3, 3, 500)
    Q_coherent = qfunc(output_1.states[index_output_data], xvec, xvec, g = 2)
    fig = plt.figure(figsize=(4, 4))
    plt.contourf(xvec, xvec, Q_coherent, 100)
    plt.text(-2.5, 2.5, 't = %5d ns'% time_point, horizontalalignment='left',verticalalignment='center',\
         fontsize=12, color='white')
    plt.savefig('./frames_Q/%04d.png'%(index_output), transparent=True,
            dpi=100, bbox_inches='tight')
    plt.close(fig)
```

```python
tot_frames = 50
for index_output in np.arange(0, tot_frames):
    index_output_data = int(index_output * 80000/tot_frames)
    time_point = 8000 * index_output/tot_frames    # ns
    xvec = np.linspace(-3, 3, 500)
    Wigner_coherent = wigner(output_1.states[index_output_data], xvec, xvec, g = 2)
    fig = plt.figure(figsize=(4, 4))
    plt.contourf(xvec, xvec, Wigner_coherent, 100)
    plt.text(-2.5, 2.5, 't = %5d ns'% time_point, horizontalalignment='left',verticalalignment='center',\
         fontsize=12, color='white')
    plt.savefig('./frames_wigner/%04d.png'%(index_output), transparent=True,
            dpi=100, bbox_inches='tight')
    plt.close(fig)
```
